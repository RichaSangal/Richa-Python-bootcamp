```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# Select rows and columns from dataframe for plotting
x = df.iloc[:,:1].values.reshape(-1,1)
y = df.iloc[:,-1].values.reshape(-1,1)


plt.scatter(x,y,color = 'b' ) 
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Fit the model and Plot the data and model
simple_model = LinearRegression()
simple_model.fit(x,y)
y_pred = simple_model.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred)
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
#Predict an unknown value
unk_x = [[70]]

x_predict = np.append(x,unk_x).reshape(-1,1)
y_predict = simple_model.predict(x_predict)

plt.scatter(x,y,color = 'b')
plt.plot(x_predict,y_predict, color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```
